### Practice visualizations using matplotlib and seaborn.

```python
# import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.to_csv("iris.csv")
```

```python
# Bar plot
sns.barplot(x = "species" , y = "sepal_length" ,hue="species",data=df)
```

```python
# Count plot
sns.countplot(x="petal_width" , data=df , hue="species")
```

```python
# Box Plot
sns.boxplot(data= df , x ="petal_width", y="petal_length" , palette="Set2")
```

```python
sns.boxenplot(data= df , x ="petal_width", y="petal_length" , hue="species")
```

```python
sns.pairplot(data=df)
```

```python
# Scatter plot 
plt.scatter(data=df , x = "sepal_width" , y="sepal_length" , c="black")
```

```python
sns.scatterplot(data=df , x="petal_length" , y="sepal_width" , hue="species" , palette=["red" , "blue" , "orange"])
```

```python
# Histogram Plot

df.hist(bins=10 , figsize=(9,7) , grid=False , color = "green" )
```

```python
# strip plot
sns.stripplot(data=df  , x="petal_length" , y = "sepal_length" , hue="species")
```

```python
# heat plot

corr = df.select_dtypes(include=["number"]).corr()
plt.figure(figsize=(7 , 7))
sns.heatmap(corr)
plt.title("Heat Plot of Iris Data Set")
```
